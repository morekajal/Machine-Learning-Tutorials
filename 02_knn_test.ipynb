## KNN Test file with Iris Dataset

```python
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
cmap = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])
```

## Import Data

```python
iris = datasets.load_iris()
X, y = iris.data, iris.target
```

```python
print(len(X), len(y))
```

## Split data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1234)
```

```python
## Lets see How Training samples looks
print(X_train.shape)        #np array of 120*4, 120->no.of samples, 4->no.of features per sample
print(X_train[0])
```

```python
## Lets look at Training labels
print(y_train.shape)       # 120
print(y_train)             # 1D row vector of size 120
```

### We can see labels 0, 1, 2 that means it is a Three Class Problem

### Let's Plot some features

```python
plt.figure()
plt.scatter(X[:, 0], X[:,1], c = y, cmap=cmap, edgecolor = 'k', s = 30)
plt.show()
```

```python
from knn import KNN
# create a classifier
clf = KNN(k = 3)
clf.fit(X_train, y_train)
predictions = clf.predict(X_test)

acc = np.sum(predictions == y_test) / len(y_test)
print(acc)
```
